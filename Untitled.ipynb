```python
import os
os.chdir(r'C:\Users\behna\OneDrive\Documents\Data Science - Projects\20210608 StackOverflow\2. Prepared Data')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn

from segtok.tokenizer import web_tokenizer
from copy import deepcopy

df = pd.read_csv(r'train.csv')
df = df.iloc[0:501]

test = df['Title'][0:2]
document = test.apply(web_tokenizer)

# building counter to track every new word coming in
global_counter = []
for sentence in range(len(document)):
    for i in range(len(document[sentence])):
        if document[sentence][i] not in global_counter:
            global_counter.append(document[sentence][i])

# assigning index value to each word and adding to dictionary
s2i = {}
for i in range(len(global_counter)):
        s2i[global_counter[i]] = i
    
# =============================================================================
# alternative counting mthod!!
# s2i = {}
# for sentence in range(len(words)):
#     for i in range(len(words[sentence])):
#         if words[sentence][i] not in s2i.keys():
#             s2i[words[sentence][i]] = len(s2i)
# =============================================================================
# prevent reference overwrite
documents = document.apply(deepcopy)

# vectorizing strings
vectors = []
for sentence in document:
    vectors.append(sentence)
    for index, word in enumerate(vectors[-1]):
        if vectors[-1][index] in s2i.keys():
            vectors[-1][index] = s2i[word]


# =============================================================================
# Building functions to apply on df
# =============================================================================
def string2int(df, target: str):
    df1 = df.apply(deepcopy)
    df1[target] = df1[target].apply(web_tokenizer)
    global_counter = []
    for sentence in range(len(df1[target])):
        for i in range(len(df1[target][sentence])):
            if df1[target][sentence][i] not in global_counter:
                global_counter.append(df1[target][sentence][i])
    
    s2i = {}
    for i in range(len(global_counter)):
            s2i[global_counter[i]] = i+1 # start from1, reserve 0 for padding
            
    df2 = df1.apply(deepcopy)
    vectors = []
    for sentence in df2[target]:
        vectors.append(sentence)
        for index, word in enumerate(vectors[-1]):
            if vectors[-1][index] in s2i.keys():
                vectors[-1][index] = s2i[word]
    return vectors

df['word_vectors'] = string2int(df, 'Title')

df_mod = df[['word_vectors', 'Y']]
df_mod.columns
df_mod

# identify longest sentence and padd all other sentences to the same length
list_len = []
for i in df['word_vectors']:
    list_len.append(len(i))

print(max(list_len))

# extending list/setnece with padding
for i in df_mod['word_vectors']:
    if len(i) < 33:
        missing_padds = 33 - len(i)
        i.extend([0]*missing_padds)
#    list_len = []
#    list_len.append(len(i))
#    print(set(list_len)) #check maximum values in list


# ordinal encoding of string to values
encoding_dict = {'LQ_CLOSE': 1, 'HQ': 2, 'LQ_EDIT': 3}
df_mod['Y'] = df_mod['Y'].map(encoding_dict)

# split and predict data
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

X = df_mod['word_vectors']

# conert column of lists to matrix for modeling
# 1 row is 33 columns
X = np.array(X.tolist())

y = df_mod['Y']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7, stratify=y)

# check shapes for feature dimensions
X_train.shape
X_test.shape
```
